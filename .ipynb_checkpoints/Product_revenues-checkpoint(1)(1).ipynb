```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import mpl
# 在python脚本中动态设置matplotlib字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
```

```python
data  = pd.read_excel('All_product_information.xlsx')
```

```python
data.columns
```

```python
data.drop(['新编码','新规格型号','地区'],axis = 1, inplace= True)
```

```python
# 重新命名特定列名
data.rename(columns = {'年':"销售时间"},inplace = True)
```

```python
# 查看大类是否均为支架并进行合计
np.round(data.groupby(["销售时间","大类"]).sum(),0)
```

```python
# 根据生产厂家分类
np.round(data.groupby(["销售时间","生产厂家"]).sum(),0)
```
