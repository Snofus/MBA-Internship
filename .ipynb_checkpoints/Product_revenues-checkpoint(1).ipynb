```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data  = pd.read_csv("product_revenue.csv",encoding = "utf-8",delimiter = ",")
```

```python
data.head()
```

```python
data.columns
```

```python
# 重新命名特定列名
data.rename(columns = {'年':"销售时间"},inplace = True)
```

```python
# 查看大类是否均为支架并进行合计
np.round(data.groupby(["销售时间","大类"]).sum(),0)
```

```python
# 根据生产厂家分类
np.round(data.groupby(["销售时间","生产厂家"]).sum(),0)
```
