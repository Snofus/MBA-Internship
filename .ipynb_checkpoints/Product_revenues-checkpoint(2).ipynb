```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import mpl
# 在python脚本中动态设置matplotlib字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
```

```python
data  = pd.read_excel('All_product_information.xlsx')
```

```python
data.columns
```

```python
data.drop(['新编码','新规格型号','地区'],axis = 1, inplace= True)
```

```python
# 重新命名特定列名
data.rename(columns = {'年':"销售时间"},inplace = True)
```

```python
data.head()
```

```python
# 查看大类是否均为支架并进行合计

equip_categories = data[['销售时间','大类','销售数量', '销售含税总价', '采购含税总价', '采购含税单价', '销售含税单价', '返利金额', '总毛利']].groupby(['销售时间','大类']).sum()
equip_categories['毛利率'] = equip_categories['总毛利']/equip_categories['销售含税总价']
equip_categories['采购含税单价'] = equip_categories['采购含税总价']/equip_categories['销售数量']
equip_categories['销售含税单价'] = equip_categories['销售含税总价']/equip_categories['销售数量']
equip_categories = np.round(equip_categories,0)
equip_categories
```

```python
# 大类手术销售曲线图
equip_categories['销售含税总价'].unstack()[['先心','支架','球囊','电生理','起搏器']].plot(figsize = (12,8))
plt.gca().set_yticklabels(['0','0','20m','40m','60m','80m','100m'])
plt.legend(['CHD',"DES","PTCA","Electrophysiology",'Pacemaker'],loc=(1.04,0.5))
plt.ylabel('Total Revenue(RMB)')
plt.xlabel('Years')
```

### Stents

```python
stent = data[data['大类']=='支架'].copy()
```

```python
stent.where()
```

```python
stent_groupby_client = stent[['生产厂家',"销售时间","客户名称"]].copy()
stent_groupby_client.rename(columns = {'生产厂家':"数量"},inplace = True)
stent_groupby_client = stent_groupby_client.groupby(['销售时间','客户名称']).count()
stent_groupby_client[stent_groupby_client['数量'] > 50]
```

```python
# Categorized by time period and manufacturer
stent_groupby_manufacturer = stent.groupby(["销售时间","生产厂家"]).sum()
stent_groupby_manufacturer.rename(columns = {'销售含税单价':'销售含税平均单价'},inplace = True)
stent_groupby_manufacturer['毛利率'] = stent_groupby_manufacturer['总毛利']/stent_groupby_manufacturer['销售含税总价']
stent_groupby_manufacturer['销售含税平均单价'] = stent_groupby_manufacturer['销售含税总价']/stent_groupby_manufacturer['销售数量']
stent_groupby_manufacturer['采购含税单价'] = stent_groupby_manufacturer['采购含税总价']/stent_groupby_manufacturer['销售数量']
stent_groupby_manufacturer = np.round(stent_groupby_manufacturer,2)
stent_groupby_manufacturer
```

```python
stent_sales_manu = stent_groupby_manufacturer['销售含税总价'].unstack()
stent_sales_manu
```

```python
# stent revenues generated by different brands over three-year period
stent_sales_manu.plot(figsize = (12,8))
plt.gca().set_yticklabels(['0','0','10m','20m','30m','40m','50m','60m','70m'])
plt.legend(['MicroPort','LePu','JWMS','Essen','Boston Scientific Corporation','Abbott Vascular'],loc=(1.04,0.5))
plt.ylabel('Total Revenue(RMB)')
plt.xlabel('Years')
```

```python
stent_sales_quantities_manu = stent_groupby_manufacturer['销售数量'].unstack()
```

```python
# sales quantities of stents produced by various manufacturers over three-year period
stent_sales_quantities_manu.plot(figsize = (12,8))
plt.legend(loc=(1.04,0.5))
plt.ylabel('销售个数')
```

```python
# 注意：未含销售返利
stent_avg_price = pd.DataFrame(np.round(stent_groupby_manufacturer['销售含税总价']/stent_groupby_manufacturer['销售数量'],-2),columns = ['平均售价(元)'])
stent_avg_price
```

```python
# 平均毛利率
stent_gross_margin = stent_groupby_manufacturer['毛利率'].unstack()
stent_gross_margin.rename(columns = {'上海微创医疗器械(集团)有限公司':'MicroPort',
                                     '乐普(北京)医疗器械股份有限公司':'LePu','山东吉威医疗制品有限公司':'JWMS',
                                     '易生科技(北京)有限公司':'Essen','波士顿科学公司 Boston Scientific Corporation':'BSC',
                                    '雅培心血管Abbott Vascular':'Abbott Vascular'})
```

### 球囊（PTCA）

```python
ptca = data[data['大类']=='球囊']
ptca
```

```python
# Categorized by time period and manufacturer
ptca_groupby_manufacturer = np.round(ptca.groupby(["销售时间","生产厂家"]).sum(),0)
ptca_groupby_manufacturer['毛利率'] = ptca_groupby_manufacturer['总毛利']/ptca_groupby_manufacturer['销售含税总价']
ptca_groupby_manufacturer.rename(columns = {'销售含税单价':'销售含税平均单价'},inplace = True)
ptca_groupby_manufacturer['销售含税平均单价'] = ptca_groupby_manufacturer['销售含税总价']/ptca_groupby_manufacturer['销售数量']
ptca_groupby_manufacturer['采购含税单价'] = ptca_groupby_manufacturer['采购含税总价']/ptca_groupby_manufacturer['销售数量']
ptca_groupby_manufacturer = np.round(ptca_groupby_manufacturer,2)
ptca_groupby_manufacturer = ptca_groupby_manufacturer[ptca_groupby_manufacturer['销售含税总价']>1000000]
ptca_groupby_manufacturer
```

- 采购和销售单价波动很大

```python
ptca_groupby_manufacturer['销售含税总价'].unstack().plot(figsize = (12,8))
plt.gca().set_yticklabels(['0','0','5m','10m','15m','20m','25m','30m'])
plt.legend(['Braun Melsungen','BSC','Terumo Coroporation','Abbott Vascular','LePu','APT Medical','Yinyi Biotech'],loc=(1.04,0.5))
plt.ylabel('Total Revenue(RMB)')
plt.xlabel('Year')
```

```python
ptca_groupby_manufacturer['销售数量'].unstack().plot(figsize = (12,8))
plt.legend(['Braun Melsungen','BSC','Terumo Coroporation','Abbott Vascular','LePu','APT Medical','Yinyi Biotech'],loc=(1.04,0.5))
plt.ylabel('销售数量')
```

```python
# Gross margin of ptca generated by different manufacturers
ptca_gross_margin = ptca_groupby_manufacturer['毛利率'].unstack()
ptca_gross_margin.rename(columns = {'乐普(北京)医疗器械股份有限公司':'LePu','湖南埃普特医疗器械有限公司':'API Medical',
                                    '辽宁垠艺生物科技股份有限公司':'Yinyi Biotech',"波士顿科学公司 Boston Scientific Corporation":'Boston Scientific Corporation',
                                   "泰尔茂株式会社 Terumo Corporation(テルモ株式会社)":'Terumo Corporation',
                                   '雅培心血管Abbott Vascular':"Abbott Vascular"})
```
